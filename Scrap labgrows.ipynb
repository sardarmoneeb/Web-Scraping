```python
import os
import time
import csv  # Import CSV module
from selenium import webdriver
from selenium.webdriver.common.by import By

# Ensure your WebDriver path is correctly set
webdriver_path = r"C:/Users/PMLS/Desktop/My program/Web Scraping/WebdriverSelenium"
os.environ['PATH'] += webdriver_path

# Initialize Chrome WebDriver
driver = webdriver.Chrome()

# Prepare a list to store the data for all pages
ring_data = []

# Loop through 20 pages
for page in range(1,21):  # Pages from 1 to 20
    driver.get(f"https://au.grownbrilliance.com/engagement-rings?page={page}&utm_campaign=oth_r&utm_source=www.grownbrilliance.com&utm_medium=wi_proxy&utm_content=en_AU&utm_term=c")
    time.sleep(5)  # Wait for the page to load

    # Loop to scrape data on the current page
    for i in range(1, 23):  # Assuming 22 items per page
        try:
            # Extract name using CSS selector
            name = driver.find_element(By.CSS_SELECTOR, f"#productGrid > div:nth-child({i}) > div > div.nproduct_name.designer_name > a").text

            # Extract price using CSS selector
            price = driver.find_element(By.CSS_SELECTOR, f"#productGrid > div:nth-child({i}) > div > div.ss__price--flex > div.nproduct_price.product_our_price").text

            # Append name and price to the data list
            ring_data.append([name, price])

            # Print to console for debugging
            print(f"Name: {name}")
            print(f"Price: {price}")
            print('-' * 50)

        except:
            # print(f"Error scraping item {i} on page {page}: {e}")
            continue

# Save all data to a CSV file
with open('All_engagement_rings.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)

    # Write header row
    writer.writerow(["Name", "Price"])

    # Write all data rows
    writer.writerows(ring_data)

print("Data saved to 'All_engagement_rings.csv'")

# Close the browser
driver.quit()
```
