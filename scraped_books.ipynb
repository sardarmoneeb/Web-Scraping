```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Base URL pattern (adjusted for correct URL format)
base_url = "http://books.toscrape.com/catalogue/page-{}.html"  # Adjusted to match the page format

# Initialize an empty list to store all the scraped data
all_data = []

# Define the number of pages to scrape (set to 20)
total_pages = 20  # Now scraping 20 pages

# Loop through each page
for page_num in range(1, total_pages + 1):
    # Construct the full URL for the current page
    page_url = base_url.format(page_num)
    print(f"Scraping page {page_num}: {page_url}")
    
    # Send a GET request to the current page
    response = requests.get(page_url)
    
    # Check if the request was successful
    if response.status_code == 200:
        print(f"Successfully retrieved page {page_num}")
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Find all the li tags containing book data on this page
        li_tags = soup.find_all('li')
        
        # Debugging: check if we are retrieving any li tags
        if not li_tags:
            print(f"Warning: No 'li' tags found on page {page_num}")
        
        # Loop through each li tag and extract book data
        for li in li_tags:
            book_tag = li.find('a')  # Find the anchor tag
            book_name = "N/A"  # Default value for book name
            
            # Check if the anchor tag is present and extract book name
            if book_tag:
                img_tag = book_tag.find('img')  # Find the img tag within the anchor
                if img_tag and img_tag.has_attr('alt'):
                    book_name = img_tag['alt']  # Extract the alt attribute as the book title
                else:
                    book_name = "N/A"  # If no book name found in the alt attribute
            
            # Extract price and availability
            price_tag = li.find(class_='product_price')
            availability_tag = li.find(class_="instock availability")
            
            # If price or availability is missing, set it to "N/A"
            price = price_tag.get_text() if price_tag else "N/A"
            availability = availability_tag.get_text() if availability_tag else "N/A"
            
            # Append the data to the list
            all_data.append({
                "book_name": book_name,
                "price": price,
                "availability": availability,
            })
    else:
        print(f"Failed to retrieve page {page_num} (Status Code: {response.status_code})")

# Convert the list of data into a DataFrame
df = pd.DataFrame(all_data)

# Save the DataFrame to an Excel file
df.to_excel("scraped_books_data_20_pages_corrected_debug.xlsx", index=False)

print("Data saved to scraped_books_data_20_pages_corrected_debug.xlsx")
```
